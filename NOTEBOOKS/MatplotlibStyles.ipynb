# Matplotlib/Seaborn Style gallery
Matplotlib and seaborn have many builtin predefined styles. This notebook plots the same data using each different style.

## Required imports

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

## Creating the data to plot
Create some cosine data to plot

```python
x_data = np.arange(0.0, 10.0, 0.1)
y_data = np.cos(x_data)
```

## Matplotlib styles
Matplotlib has several builtin styles, and seaborn adds many more. You can also define your own matplotlib stylesheets. 

```python
plt.style.available
```

## Selecting a style
To select a style, use `plt.style.use('STYLENAME')`. All subsequent plots will use this style, unless you call `plt.style.use()` with a different style. 

```python
plt.style.use('seaborn-v0_8-pastel')
plt.plot(x_data, y_data)
```

```python
plt.style.use('ggplot')
plt.plot(x_data, y_data)
```

```python
with plt.style.context('seaborn-v0_8-pastel'):
    plt.plot(x_data, y_data)
```

## Temporary styling
the `plt.style.context()` function is a context manager that will use a style temporarily, just within the scope of the `with` block. 

```python
with plt.style.context('fivethirtyeight'):
    titles = ['alpha', 'beta', 'gamma'] 

    y = np.arange(1, 11)
    for title in titles:
        data = np.random.random_sample((10,)) * 10
        plt.plot(y, data, label=title)
    plt.ylim(0, 15)
    plt.legend(loc="upper right")
    plt.text(x=3, y=12, weight='bold', s="Cool Data", fontsize=24)
    plt.text(x=3, y=11, weight='bold', s="Random Integers", fontsize=16)
    plt.show()
```

## Reverting to default style
To revert to the default style, use the style conveniently named 'default'.

```python
plt.style.use('default')
```

## Style Gallery
Loop through all available styles and plot the data.

```python
mpl.rcParams['figure.max_open_warning'] = 30  # allow more than 20 figures without a warning
for style_name in mpl.style.available:
    if style_name.startswith('_'):
        continue  # skip "private" styles
    plt.figure()
    with mpl.style.context(style_name):
        plt.title(style_name)
        plt.plot(x_data, y_data)
```
